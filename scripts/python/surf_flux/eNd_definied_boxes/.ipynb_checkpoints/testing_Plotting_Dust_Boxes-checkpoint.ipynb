# A script to plot a map of the boxes of dust flux with land sea mask

## Loading the netCDF file with the distinct boxes in the ocean and the land-sea mask 

```python
# load packages

%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import matplotlib.gridspec as gridspec # subplots
from matplotlib import colors as mcolors

import iris
import iris.plot as iplt
import iris.quickplot as qplt
#plt.interactive(True)

import numpy as np
import pandas as pd

import cf_units as unit

#iris.FUTURE.netcdf_promote=True
```

```python

fname = '/nfs/see-fs-01_users/ee14s2r/working/surf_files/eNd_dust_mask/dust_boxes_eNd.nc'

cube=iris.load(fname)


##print the cubes, this shows the list of cubes and their indexing 
print(cube)


# Using cube[0] 
N_Atlantic= cube[4][:,:]
Atlantic= cube[0][:,:]
N_Pacific = cube[1][:,:]
I_Pacific = cube[2][:,:]
Background = cube[3][:,:]


latitude,longitude=N_Atlantic.shape
latitude,longitude=Atlantic.shape
latitude,longitude=N_Pacific.shape
latitude,longitude=I_Pacific.shape
latitude,longitude=Background.shape

print N_Atlantic.shape
print I_Pacific.shape
```

```python
plt.figure(figsize=(20, 12))

plt.subplot(1, 2, 1)
iplt.contourf(N_Pacific[:,:])
plt.title('North Pacific Dust Region',fontsize=16)
plt.set_cmap('brewer_Dark2_08')
ax = plt.gca()
ax.coastlines()

plt.subplot(1, 2, 2)
iplt.contourf(I_Pacific[:,:])
plt.title('Indo Pacific Dust Region',fontsize=16)
plt.set_cmap('brewer_Accent_08')
ax = plt.gca()
ax.coastlines()

plt.show()
```
